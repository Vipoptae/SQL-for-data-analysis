# Group By

```python
sql='''
SELECT
Country,
count(*) as "count country"
FROM
customers
group by Country
ORDER by count(country) desc
'''
data=pd.read_sql(sql, con)
data.head()
```

# Having
where filter individual ,
having from group by 

```python
sql='''
SELECT
Country,
count(*) as "count country"
FROM
customers
group by Country
having "count country" >=4
order by "count country" desc 
'''
data=pd.read_sql(sql, con)
data.head()
```

```python
sql='''
SELECT
Country,
count(*) as "count country"
FROM
customers
group by 1 
having "count country" >=4
order by 2 '''
data=pd.read_sql(sql, con)
data.head()
```

# LIKE 
same with '=' but can use with wildcard (c% => first char with c)


%gmail.com


order of char'a' => (_a%)=> Canada


(__a%)=> Brazil

```python
sql='''
SELECT Country,
count(*) as count_country
FROM customers
where country like 'c%'
GROUP	by Country	
'''
data=pd.read_sql(sql, con)
data.head()
```

```python
sql='''
SELECT Country,
count(*) as count_country
FROM customers
where country like '_a%'
GROUP	by Country	
'''
data=pd.read_sql(sql, con)
data.head()
```
